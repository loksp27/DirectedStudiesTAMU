```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd

# Data from the provided text
data = {
    "4/2/2023": 1, "4/10/2023": 2, "4/14/2023": 1, "4/17/2023": 1, "4/18/2023": 1, "4/19/2023": 4, "4/20/2023": 4, "4/21/2023": 3, "4/24/2023": 3, "4/25/2023": 1, "4/26/2023": 1, "4/27/2023": 7, "4/28/2023": 6, "4/29/2023": 1, "5/1/2023": 1, "5/3/2023": 2, "5/5/2023": 2, "5/7/2023": 1, "5/8/2023": 1, "5/9/2023": 1, "5/10/2023": 1, "5/11/2023": 4, "5/12/2023": 1, "5/14/2023": 1, "5/15/2023": 1, "5/17/2023": 1, "5/18/2023": 1, "5/19/2023": 1, "5/20/2023": 1, "5/21/2023": 1, "5/22/2023": 1, "5/23/2023": 2, "5/24/2023": 1, "5/25/2023": 1, "5/29/2023": 6, "5/30/2023": 4, "6/1/2023": 2, "6/4/2023": 2, "6/5/2023": 3, "6/6/2023": 9, "6/7/2023": 3, "6/8/2023": 3, "6/9/2023": 2, "6/11/2023": 1, "6/12/2023": 1, "6/13/2023": 3, "6/14/2023": 2, "6/15/2023": 2, "6/16/2023": 3, "6/17/2023": 1, "6/18/2023": 3, "6/19/2023": 3, "6/20/2023": 3, "6/21/2023": 5, "6/22/2023": 4, "6/23/2023": 1, "6/26/2023": 4, "6/27/2023": 4, "6/28/2023": 6, "6/29/2023": 4, "6/30/2023": 7, "7/3/2023": 1, "7/4/2023": 2, "7/5/2023": 10, "7/6/2023": 2, "7/9/2023": 1, "7/11/2023": 1, "7/18/2023": 2
}

# Convert data to DataFrame for easier manipulation
df = pd.DataFrame(list(data.items()), columns=["Date", "Value"])

# Convert the Date column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Set the Date column as the index
df.set_index("Date", inplace=True)

# Resample the data to weekly frequency and sum the values
weekly_data = df.resample('W').sum()

# Convert to cumulative sum
cumulative_data = weekly_data.cumsum()

# Generate list of dates corresponding to the weeks
dates = cumulative_data.index.strftime('%m/%d/%Y').tolist()

# Convert the weekly cumulative bug counts to a list
Lx90_V22_2_1 = cumulative_data['Value'].tolist()

weekly_population_data = Lx90_V22_2_1[3:]
population_data = np.array(Lx90_V22_2_1[:3])
print(f"All of observed data: {Lx90_V22_2_1}")

# Initial population data for week 0
time_data = np.array([0, 1, 2])

initial_weeks = len(population_data)
milestone_week = len(population_data) + len(weekly_population_data)

# Gompertz growth function
def gompertz_model(t, K, b, r):
    return K * np.exp(-b * np.exp(-r * t))

# Richards growth function
def richards_model(t, K, r, t0, nu):
    return K * (1 + nu * np.exp(-r * (t - t0)))**(-1/nu)

# Logistic growth function
def logistic_model(t, K, P0, r):
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

# Iterate over each week's data and fit the model
for week, new_population in enumerate(weekly_population_data, start=0):
    # Add the new week's data
    time_data = np.append(time_data, week + initial_weeks)
    population_data = np.append(population_data, new_population)
    len_observed_data = len(population_data)
    print(f"len_observed_data {len_observed_data}")
    print(f"population {population_data}")

    # Fit the logistic model to the data using curve_fit
    popt, _ = curve_fit(logistic_model, time_data, population_data, p0=[100, 10, 0.1], maxfev=5000, bounds=(0, [np.inf, np.inf, np.inf]))
    K_est, P0_est, r_est = popt

    # Fit the Gompertz model to the data using curve_fit
    gompertz_popt, _ = curve_fit(gompertz_model, time_data, population_data, p0=[180, 1, 0.1],maxfev=50000, bounds=(0, [np.inf, np.inf, np.inf]))
    K_gomp, b_gomp, r_gomp = gompertz_popt

    # Fit the Richards model to the data using curve_fit
    initial_guesses = [150, 0.1, 7, 1]
    Richards_popt, _ = curve_fit(richards_model, time_data, population_data, p0=initial_guesses, maxfev=5000, bounds=(0, [np.inf, np.inf, np.inf, np.inf]))
    R_K_est, R_r_est, R_t0_est, R_nu_est = Richards_popt

    # Predict future population values using the fitted model
    time_future = np.arange(0, milestone_week + 10)
    population_future = logistic_model(time_future, K_est, P0_est, r_est)
    gompertz_prediction = gompertz_model(time_future, K_gomp, b_gomp, r_gomp)
    R_time_future = np.arange(0, milestone_week + 10, 1)
    richards_prediction = richards_model(R_time_future, R_K_est, R_r_est, R_t0_est, R_nu_est)

    # Skip if data length is too short or K_est is negative
    if len(time_data) <= 3 or K_est < 0 or len_observed_data < 9:
        print(f"Do Not print anything: {len_observed_data}")
        continue

    # Ensure dates length matches time_future length
    while len(dates) < len(time_future):
        dates.append((pd.to_datetime(dates[-1]) + pd.DateOffset(weeks=1)).strftime('%m/%d/%Y'))
    date_labels = dates[:len(time_future)]

    # Calculate MSE for the entire dataset
    mse_logistic = np.mean((population_future[:len(Lx90_V22_2_1)] - Lx90_V22_2_1[:len(population_future)])**2)
    mse_gompertz = np.mean((gompertz_prediction[:len(Lx90_V22_2_1)] - Lx90_V22_2_1[:len(gompertz_prediction)])**2)
    mse_richards = np.mean((richards_prediction[:len(Lx90_V22_2_1)] - Lx90_V22_2_1[:len(richards_prediction)])**2)

    # Calculate MSE for each week
    mse_weeks = [(i, (population_future[i] - Lx90_V22_2_1[i])**2) for i in range(len(population_data))]
    for week_idx, mse in mse_weeks:
        print(f"Week {week_idx + 1} MSE: {mse}")

    # Print the table of predicted populations with differences and observed values
    observed_values = np.append(population_data, [np.nan] * (len(time_future) - len(population_data)))
    original_values = np.append(Lx90_V22_2_1, [np.nan] * (len(time_future) - len(Lx90_V22_2_1)))
    prediction_table = pd.DataFrame({
        'Week': date_labels,
        'Original': original_values,
        'Observed': observed_values,
        'Logistic': population_future,
        'Gompertz': gompertz_prediction,
        'Richards': richards_prediction,
    })

    print(f"{len_observed_data} Weeks - Updated Predictions:")
    print(f"Milestone week is: {milestone_week}")
    print(prediction_table)

    # Print final estimated parameters
    print(f"Final MSE (Logistic): {mse_logistic}")
    print(f"Final MSE (Gompertz): {mse_gompertz}")
    print(f"Final MSE (Richards): {mse_richards}")
    print(f"Final estimated carrying capacity (Logistic): {K_est}")
    print(f"Final estimated growth rate (Gompertz b): {K_gomp}")
    print(f"Final estimated carrying capacity (Richards): {R_K_est}")
    print(f"Final estimated initial population (P0): {P0_est}")
    print(f"Final estimated growth rate (r): {r_est}")

    # Optionally, plot the data and predictions
    plt.figure(figsize=(15, 6))
    plt.scatter(time_data, population_data, label='Observed Data')
    plt.plot(time_future, population_future, label='Fitted Logistic Curve', color='red')
    plt.plot(time_future, gompertz_prediction, label='Gompertz Model', color='blue')
    plt.plot(R_time_future, richards_prediction, label='Richards Model', color='green')
    plt.xlabel('Date')
    plt.ylabel('Population')
    plt.title(f'Population Growth Prediction - Week {len_observed_data}')
    plt.legend()
    plt.grid(True)
    plt.xticks(ticks=range(len(date_labels)), labels=date_labels, rotation=45)
    plt.tight_layout()
    plt.show()

    print(f"The End of {len_observed_data} Weeks - Updated Predictions:")
    print(f"-----------------------------------------------------------")
```
